```python
#Import the lirbaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the data which is from the Kaggle for study purpose
data=pd.read_csv('Classified Data.txt', index_col=0) # here the index_col=0 becuase the first row in the file is column names
```

```python
# Lets divide the model into features and label
features=data.iloc[:,:10].values
label=data.iloc[:,-1].values
```

```python
features
```

```python
label
```

#### Lets divide the data into train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(features,
                                                 label,
                                                 test_size=0.2,
                                                 random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# Lets build the KNN classifier

from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier()
model.fit(X_train, y_train)
```

```python
#Keeping the parameters of the model as default, so K value will be 5


trainScore=model.score(X_train, y_train)
testScore=model.score(X_test, y_test)

print(f"Test Score: {testScore} and Train Score: {trainScore}")
```

```python
model_predict=model.predict(X_test)
```

```python
# Lets put the metrics
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
score=accuracy_score(model_predict, y_test)
print("---Score---")
print(score)
print("---Classification-Report---")
print(classification_report(model_predict, y_test))
print("---Confusion-Matrix---")
print(confusion_matrix(model_predict, y_test))
```

```python
# Lets see if by selecting Random State can we get the score more than 95.

CL=0.95

for seed in range (1,100):
    X_train, X_test, y_train, y_test=train_test_split(features,
                                                     label,
                                                     test_size=0.2,
                                                     random_state=seed)

    model2=KNeighborsClassifier()
    model2.fit(X_train, y_train)

    trainScore2=model2.score(X_train, y_train)
    testScore2=model2.score(X_test, y_test)

    if testScore2 > trainScore2 and testScore2 >= CL:
        print(f"Test Score: {testScore2} and Train Score: {trainScore2} at Ranomd State: {seed}")
```

#### I am getting the best score at random state 31
